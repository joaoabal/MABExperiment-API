## Modelagem - Rascunho Thompson Sampling

```python
samples = []
alpha = 25 # clicks
beta = 12 #non-clicks
samples.append(('A', np.random.beta(alpha, beta)))

alpha = 29 # clicks
beta = 12 #non-clicks
samples.append(('B', np.random.beta(alpha, beta)))
samples
```

```python
total = sum(sample[1] for sample in samples)
total
```

```python

allocations = {variant_name: round((sample / total) * 100, 2) for variant_name, sample in samples}
allocations
```

```python
from scipy.stats import beta

alpha_values = [10, 10, 20, 40] #clicks
beta_values = [10, 20, 20, 20] #non-clicks

x = np.linspace(0, 1, 1000)

fig, ax = plt.subplots(figsize=(10, 6))

for alpha, beta_val in zip(alpha_values, beta_values):
    y = beta.pdf(x, alpha, beta_val)
    label = f'Clicks={alpha}, Non-Clicks={beta_val}, CTR = {round(alpha/(alpha+beta_val),2)}'
    ax.plot(x, y, label=label)

ax.set_title('Distribuies Beta para Diferentes Valores de Alpha (Clicks) e Beta (Non-Clicks)')
ax.set_xlabel('Probabilidade')
ax.set_ylabel('Densidade de Probabilidade')
ax.legend()

plt.show()
```

## Analise de resultados

```python
def get_db_connection():
    conn = psycopg2.connect(
        host="localhost",               # Nome do servio do banco de dados no docker-compose
        database="mab_db",       # Nome do banco de dados
        user="postgres",         # Usurio do PostgreSQL
        password="password",     # Senha do PostgreSQL
        port=5433                # Porta do banco de dados
    )
    return conn
```

```python
def get_allocations(conn):
    query = """
        SELECT allocation_date, variant_name, allocation_percentage, start_date, end_date
        FROM allocations
        ORDER BY allocation_date;
    """
    return pd.read_sql_query(query, conn)
```

```python
def get_experiment_data(conn):
    query = """
        SELECT name, timestamp, impressions, clicks,
        (clicks::float / NULLIF(impressions, 0)) * 100 AS ctr  -- Calcular CTR, evitando diviso por zero
        
        FROM experiment_data
        ORDER BY timestamp;
    """
    return pd.read_sql_query(query,conn)
```

---

## Extrao dos dados

```python
# Conectar ao banco de dados
conn = get_db_connection()
```

```python
allocations = get_allocations(conn)
allocations
```

```python

# Buscar os dados do experimento para o intervalo
experiment_data = get_experiment_data(conn)
experiment_data

```

----

## Plot

```python
# Iterar sobre cada experimento baseado na allocation_date
for allocation_date in allocations['allocation_date'].unique():
    
    # Filtrar os dados de alocao para a current allocation_date
    current_allocation = allocations[allocations['allocation_date'] == allocation_date]
    
    # Filtrar os dados de experiment_data com base no start_date e end_date do experimento atual
    start_date = current_allocation.iloc[0].start_date
    end_date = current_allocation.iloc[0].end_date
    
    experiment_data_filtered = experiment_data.copy()
    experiment_data_filtered = experiment_data_filtered[
        (experiment_data_filtered["timestamp"] >= start_date) &
        (experiment_data_filtered["timestamp"] <= end_date)
    ]
    experiment_data_filtered = experiment_data_filtered[experiment_data_filtered["name"].isin(current_allocation.variant_name.unique())]

    # Criar grfico para o experimento atual
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Definir um mapeamento de cores para manter a coerncia
    color_map = {
        'A': 'tab:blue',
        'B': 'tab:green',
        'C': 'tab:red'
    }

    # Plotar o CTR no eixo esquerdo
    ax1.set_xlabel('Data')
    ax1.set_ylabel('CTR (%)')

    # Plotando CTR com cores consistentes para cada variante
    for variant in experiment_data_filtered['name'].unique():
        variant_data = experiment_data_filtered[experiment_data_filtered['name'] == variant]
        
        # Plotar CTR para cada variante
        ax1.plot(variant_data['timestamp'], variant_data['ctr'], 
                 label=f'CTR {variant}', 
                 color=color_map.get(variant, 'tab:blue'))

    ax1.tick_params(axis='y')

    # Configurar o segundo eixo (direito) para alocao
    ax2 = ax1.twinx()
    ax2.set_ylabel('Alocao (%)')

    # Plotar alocao como um ponto nico por variante
    for variant in current_allocation['variant_name'].unique():
        variant_allocation = current_allocation[current_allocation['variant_name'] == variant]
        
        # Colocar um ponto no grfico para a alocao
        ax2.scatter(variant_allocation['end_date'], 
                    variant_allocation['allocation_percentage'], 
                    label=f'Alocao {variant}', 
                    color=color_map.get(variant, 'tab:red'))

    ax2.tick_params(axis='y')

    # Ajustar ttulos e legendas
    fig.tight_layout()
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper center')
    
    plt.title(f'Srie Temporal de CTR e Alocaes ({allocation_date})')
    
    # Mostrar o grfico
    plt.show()
```
